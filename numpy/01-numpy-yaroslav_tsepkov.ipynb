```python
def genOlap(nday, nproduct, nsale) -> np.ndarray:
    day = []
    for _ in range(nday):
        products = []
        for _ in range(nproduct):
            step = np.random.randint(100, 1000)
            products.append(np.random.randint(step, step*10, size=(nsale,)))
        products = np.vstack(products)
        day.append(products)
    return np.dstack(day)
```

```python
olap = genOlap(3,3,3)
print(f'olap cube:\n {olap}')
```

```python
sample = olap[:3,:,1:3]
print(f'sample:\n{sample}\nmin axis 1 and 2 fixed:\n{sample.min(axis=(1,2))}')
```
