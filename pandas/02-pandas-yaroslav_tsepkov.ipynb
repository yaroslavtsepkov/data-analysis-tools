```python
for _ in range(0, 25):
    df.iloc[np.random.randint(0,df.shape[0]),1] = np.NaN # Добавим неизветсных в данные в случайную строку в колонку Open
```

```python
df.isna().sum() # Добавилось 
```

```python
df['Open'] = df['Open'].fillna(value=df['Open'].mean()) # Заменим неизвестные средним значением этой колонки 
```

```python
df.isna().sum() # Проверим что теперь снова все чисто
```

```python
df['CloseOpen'] = df['Close'] - df['Open'] # Разница между ценами на моменты закрытия и открытия торгов
```

```python
df.head()
```

```python
df['drop/growth'] = df['CloseOpen'].apply(lambda x: 'drop' if x < 0 else 'growth' ) # Применение к PandasSeries
```

```python
df['spread']=df.apply(lambda x: x['High']-x['Low'], axis=1) # Применение к PandasFrame
```

```python
statistics = ['mean', 'max', 'min', 'std', 'median']
columns = list(df)
for col in columns[1:5]:
    print(f'for {col}')
    for stat in statistics:
        print(f' {stat} = {df[col].agg(stat)} ')
```

```python
# добавим к котировкам SPCE котировки Tesla 
# но прежде чем все добавлять мы изменить тип данных Date с object на Datetime и по ней будем соединять
tsla_df = pd.read_csv(r'TSLA.csv')
tsla_df.head()
tsla_df['Date'] = pd.to_datetime(tsla_df['Date'], format='%Y-%m-%d')
tsla_df.dtypes
```

```python
spce_df = df.reset_index()
spce_df['Date'] = pd.to_datetime(spce_df['Date'])
spce_df.dtypes
```

```python
df = pd.merge(left = spce_df, right=tsla_df, how='inner', on='Date', suffixes=('_spce', '_tsla')) # соеденим данные
df = df.drop(columns=['CloseOpen', 'drop/growth', 'spread']) # уберем то что сейчас не нужно
df.head()
```

```python
# Вырежем кусок нашего фрейма чтобы потом его снова добавить, 
# конечно не самое логичное что можно сделать, ну главное показать что можно и так 
pie_df = df.iloc[20:25, :]
pie_df.head()
```

```python
print(f'before shape: {df.shape}')
df = pd.concat([pie_df, df])
print(f'after concat shape: {df.shape}')
df = df.append(pie_df)
print(f'after append shape: {df.shape}')
```
