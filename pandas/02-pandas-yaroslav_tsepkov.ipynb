```python
for _ in range(0, 25):
    df.iloc[np.random.randint(0,df.shape[0]),1] = np.NaN # Добавим неизветсных в данные в случайную строку в колонку Open
```

```python
df.isna().sum() # Добавилось 
```

```python
df['Open'] = df['Open'].fillna(value=df['Open'].mean()) # Заменим неизвестные средним значением этой колонки 
```

```python
df.isna().sum() # Проверим что теперь снова все чисто
```

```python
df['CloseOpen'] = df['Close'] - df['Open'] # Разница между ценами на моменты закрытия и открытия торгов
```

```python
df.head()
```

```python
df['drop/growth'] = df['CloseOpen'].apply(lambda x: 'drop' if x < 0 else 'growth' ) # Применение к PandasSeries
```

```python
df['spread']=df.apply(lambda x: x['High']-x['Low'], axis=1) # Применение к PandasFrame
```

```python
df.sort_values(by='Open')
```

```python
statistics = ['mean', 'max', 'min', 'std', 'median']
columns = list(df)
for col in columns[1:5]:
    print(f'for {col}')
    for stat in statistics:
        print(f' {stat} = {df[col].agg(stat)} ')
```
